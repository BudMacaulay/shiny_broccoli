```python
# - - - Math crunching to yield what colour I guess - - - #

# Assume that the prefentiallity of Mn in a Mn system is 0 (cause obviously)

# Must do Surface state energy - bulk state energy = prefential to surface
# Implies a negative value means that said species prefers to be on the surface for the said bulk (more so than the pure MO system anyway)
# Positive value - - said species would prefer to be in the bulk.

# Since there are so little runs seems a bit over kill to pull the results into the notebook. - I'll instead manually pull 'em
# Will still use the vasprun stuff and my excel thingy to put them in a decen format. 
# Key note the systems that are (34,2,0) are stio. equivalent to those that are (17,1,0) and can therefore be directly compared. 
# Perhaps doing some goofy chemical potential stuff for the 1,3,1 systems could be interesting

# - - - Mn subs - - - #

    ## IN Co ##
Co4Mnbulktot = -406.1093122
Co4MnbE = Co4Mnbulktot / 72
Co4Mnsurftot = -817.1871594
Co4MnsE = Co4Mnsurftot / 144
MninCo_pref = Co4MnsE - Co4MnbE
print(colored("Mn's surface prefentially is " + str(1000 * MninCo_pref) + "meV in Co",'blue'))

    ## IN Ni ##
Ni4Mnbulktot = -363.7922187
Ni4MnbE = Ni4Mnbulktot / 72
Ni4Mnsurftot = -724.2012989
Ni4MnsE = Ni4Mnsurftot / 144
MninNi_pref = Ni4MnsE - Ni4MnbE 
print(colored("Mn's surface prefentially is " + str(1000 * MninNi_pref) + " meV in Ni" + "\n\n",'red'))

# - - - Co subs - - - #
    
    ## IN Mn ##
Mn4Cobulktot = -464.6993842
Mn4CobE = Mn4Cobulktot / 72
Mn4Cosurftot = -913.2366647
Mn4CosE = Mn4Cosurftot / 144
CoinMn_pref = Mn4CosE - Mn4CobE
print(colored("Co's surface prefentially is " + str(1000 * CoinMn_pref) + " meV in Mn","red"))
    
    ## IN Ni ##
Ni4Cobulktot = -359.8525847
Ni4CobE = Ni4Cobulktot / 72
Ni4Cosurftot = -718.4602761
Ni4CosE = Ni4Cosurftot / 144
CoinNi_pref = Ni4CosE - Ni4CobE
print(colored("Co's surface prefentially is " + str(1000 * CoinNi_pref) + " meV in Ni" + "\n\n","red"))



# - - - Ni subs - - - #
    ## IN Co ##
Co4Nibulktot = -401.2165656
Co4NibE = Co4Nibulktot / 72
Co4Nisurftot = -807.5652546
Co4NisE = Co4Nisurftot / 144
NiinCo_pref = Co4NisE - Co4NibE
print(colored("Ni's surface prefentially is " + str(1000 * NiinCo_pref) + " meV in Co",'blue'))
    ## IN Mn ##
Mn4Nibulktot = -462.1105882
Mn4NibE = Mn4Nibulktot / 72
Mn4Nisurftot = -923.7115537
Mn4NisE = Mn4Nisurftot / 144
NiinMn_pref = Mn4NisE - Mn4NibE
print(colored("Ni's surface prefentially is " + str(1000 * NiinMn_pref) + " meV in Mn","green"))

# Colour grad these?

Mn_ds = [(0,2,34,MninCo_pref),(34,2,0,MninNi_pref),(0,36,0,0)]
Mn_points = [(0,2,34),(34,2,0),(0,36,0)]
Mn_energies = [MninCo_pref, MninNi_pref, 0]

Ni_ds = [(2,0,34,NiinCo_pref),(2,34,0,NiinMn_pref),(36,0,0,0)]
Ni_points = [(2,0,34),(2,34,0),(36,0,0)]
Ni_energies = [NiinCo_pref, NiinMn_pref, 0]

Co_ds = [(0,34,2,CoinMn_pref),(34,0,2,CoinNi_pref),(0,0,36,0)]
Co_points = [(0,34,2),(34,2,0),(0,0,36)]
Co_energies = [CoinMn_pref, CoinNi_pref, 0]

Ni_colors = []
for n in Ni_energies:
    if n < 0:
        Ni_colors.append('b')
    elif n > 0: 
        Ni_colors.append('r')
    elif n == 0:
        Ni_colors.append('g')
        
Mn_colors = []
for n in Mn_energies:
    if n < 0:
        Mn_colors.append('b')
    elif n > 0: 
        Mn_colors.append('r')
    elif n == 0:
        Mn_colors.append('g')
        
Co_colors = []
for n in Co_energies:
    if n < 0:
        Co_colors.append('b')
    elif n > 0: 
        Co_colors.append('r')
    elif n == 0:
        Co_colors.append('g')
        

print(colored("\n\nPotentially showing that Co has a high tendancy to prefer the bulk relative to the other atom types? - Thoughts","magenta"))
print(colored('As observed by the positive values for Co substitions aswell as the negative values for the M2 subs in Co',"magenta"))
print(colored('could be interesting result','magenta'))

set = 2
```
